TENSOR MATH OPERATIONS
* Add, Subtract, Multiply, Divide, Remainders, Exponents
* Shorthand and uh, Longhand
* Reassignment




```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
# Addition
tensor_a + tensor_b
```

```python
# Addition Longhand
torch.add(tensor_a, tensor_b)
```

```python
# Subtraction
tensor_b - tensor_a
```

```python
# Subtraction Longhand
torch.sub(tensor_b, tensor_a)
```

```python
# Multiply Longhand
torch.mul(tensor_b, tensor_a)
```

```python
# Divide Longhand
torch.div(tensor_b, tensor_a)
```

```python
# Modulus
tensor_b % tensor_a
```

```python
# Modulus Longhand
torch.remainder(tensor_b, tensor_a)
```

```python
# Exponents / Power
pow_1 = torch.pow(tensor_a, tensor_b)
pow_2 = tensor_a ** tensor_b

pow_1, pow_2
```
